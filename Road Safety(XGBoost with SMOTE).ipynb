```python
import numpy as np
import pandas as pd
file_1= pd.read_csv('Cas.csv')
file_2= pd.read_csv('dftRoadSafety_Accidents_2016.csv')
file_3= pd.read_csv("MakeModel2016.csv")
```

```python
pre_data= file_1.merge(file_2, on="Accident_Index",how="inner")
dataset= pre_data.merge(file_3, on="Accident_Index", how="inner")
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.drop(columns='Accident_Index', axis=1, inplace=True)
```

```python
dataset.info()
```

```python
pd.set_option('display.expand_frame_repr', False) 
dataset.mask(dataset==-1, inplace=True)

```

```python
nulls = dataset.isnull().sum().to_frame()
for index, row in nulls.iterrows():
    print(index, row[0])
```

```python

dataset.dropna(how='any',inplace=True)
```

```python
dataset.info()
```

```python
# correlation matrix 
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(30,30))
sns.heatmap(dataset.corr(), annot = True)

plt.show()
```

```python
# dropping correlated and unnecessary columns for prediction
dataset.drop(columns=['make','model','accyr','Longitude','Latitude',"Age_of_Casualty",'Pedestrian_Location','Casualty_Severity','Police_Force','Casualty_Reference', \
                      'Junction_Control','2nd_Road_Class', 'Local_Authority_(Highway)', 'LSOA_of_Accident_Location','1st_Road_Class','1st_Road_Number','2nd_Road_Number'], axis=1, inplace=True)
```

```python
dataset['Date']
```

```python
# Covert 'Date' to proper datetime format
dataset['Date']= pd.to_datetime(dataset['Date'])
```

```python
# extract month from date
dataset['Month'] = pd.DatetimeIndex(dataset['Date']).month
```

```python
# extract hour from time
dataset['Hour_of_the_day'] = pd.to_datetime(dataset['Time']).dt.hour
```

```python
dataset.info()
```

```python
#  dropping time and date as they are no longer useful
dataset.drop(columns=['Time', 'Date'],axis=1, inplace=True)
```

```python
dataset.info()
```

```python
dataset['Accident_Severity'].value_counts()
```

```python
# merging '1'- fatal accidents with '2'- serious accidents and denoting both as '1'
# denoting '3'-slight accidents as 0
dataset['Accident_Severity'] = dataset['Accident_Severity'].replace(2, 1)

dataset['Accident_Severity'] = dataset['Accident_Severity'].replace(3, 0)


dataset['Accident_Severity'].value_counts()
```

# Building a Predictive Model 

```python
X= dataset.drop(columns=['Accident_Severity'])
Y= dataset.Accident_Severity
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, stratify=Y)
y_train.value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_train_df = scaler.fit_transform(x_train)
normalised_test_df = scaler.transform(x_test)

x_train = normalised_train_df
x_test = normalised_test_df
```

```python
# Oversampling to balance the imbalanced data
import imblearn
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=1)
x_train_balanced, y_balanced = smote.fit_sample(x_train, y_train)
y_balanced.value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=2, random_state=1)
clf.fit(x_train, y_train)
new_predictions = clf.predict(x_test)
```

```python
from sklearn.metrics import recall_score, accuracy_score, precision_score, f1_score, confusion_matrix
accuracy1 = accuracy_score(y_true=y_test, y_pred=new_predictions)
print('Accuracy: {}'.format(round(accuracy1*100), 2))
f1_rf = f1_score(y_test, new_predictions, average='micro')
print('f1 score: {:.2f}'.format(f1_rf))
```

```python
feature_name = X.columns

# features importance
feature_importance = pd.DataFrame(clf.feature_importances_,index=feature_name)
feature_importance = feature_importance.sort_values(0, ascending=False)
feature_importance.reset_index(level=0, inplace=True)
```

```python
 #Plot feature importance
plt.figure(figsize=(10, 10))
sns.barplot(x=feature_importance[0], y=feature_importance['index'])

plt.title('PLOT OF FEATURE IMPORTANCE FOR RANDOM FOREST')
plt.xlabel('FEATURE IMPORTANCE VALUE')
plt.ylabel('FEATURE NAMES')
```

```python
import xgboost as xgb
xg_class = xgb.XGBClassifier(objective ='binary:logistic', random_state= 1)
xg_class.fit(x_train, y_train)
new_predictions2 =xg_class.predict(x_test)
accuracy2 = accuracy_score(y_true=y_test, y_pred=new_predictions2)
print('Accuracy: {}'.format(round(accuracy2*100), 2))
f1_xgb = f1_score(y_test, new_predictions2, average='micro')
print('f1 score: {:.2f}'.format(f1_xgb))
```
